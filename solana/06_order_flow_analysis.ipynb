```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from datetime import datetime, timedelta
import warnings
from scipy import stats
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import warnings

warnings.filterwarnings('ignore')

# Set up plotting
plt.style.use('default')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (15, 10)
%matplotlib inline

print("=== BLOCK-LEVEL ORDER FLOW & FLOW TOXICITY EDA ===")
print("Objective: Explore microstructure patterns in block-level order flow and information content")
print()

# Constants
SOL_MINT = 'So11111111111111111111111111111111111111112'
DATA_PATH = Path('../data/solana/first_day_trades/first_day_trades_batch_578.csv')

# Load data
print("Loading data...")
df = pd.read_csv(DATA_PATH)
df['block_timestamp'] = pd.to_datetime(df['block_timestamp'])

# Create coin mapping
unique_mints = df['mint'].unique()
coin_names = {mint: f"Coin_{i}" for i, mint in enumerate(unique_mints, 1)}
df['coin_name'] = df['mint'].map(coin_names)

print(f"Data loaded: {len(df):,} transactions across {len(unique_mints)} coins")
print(f"Time range: {df['block_timestamp'].min()} to {df['block_timestamp'].max()}")
print()
```

```python
# Enhanced Data Processing: Add trading indicators and implicit prices
print("=== ENHANCED DATA PROCESSING ===")

# Add trading direction and SOL amounts
df['is_buy'] = df['mint'] == df['swap_to_mint']
df['is_sell'] = df['mint'] == df['swap_from_mint']
df['sol_amount'] = 0.0
df['token_amount'] = 0.0

# Calculate SOL amounts and token amounts
buy_mask = df['is_buy'] & (df['swap_from_mint'] == SOL_MINT)
sell_mask = df['is_sell'] & (df['swap_to_mint'] == SOL_MINT)

df.loc[buy_mask, 'sol_amount'] = df.loc[buy_mask, 'swap_from_amount']
df.loc[buy_mask, 'token_amount'] = df.loc[buy_mask, 'swap_to_amount']
df.loc[sell_mask, 'sol_amount'] = df.loc[sell_mask, 'swap_to_amount']
df.loc[sell_mask, 'token_amount'] = df.loc[sell_mask, 'swap_from_amount']

# Calculate implicit SOL-relative prices
# Price = SOL_amount / token_amount (SOL per token)
df['implicit_price'] = np.where(
    (df['token_amount'] > 0) & (df['sol_amount'] > 0),
    df['sol_amount'] / df['token_amount'],
    np.nan
)

print(f"Trading direction breakdown:")
print(f"  Buys: {df['is_buy'].sum():,} ({df['is_buy'].mean():.1%})")
print(f"  Sells: {df['is_sell'].sum():,} ({df['is_sell'].mean():.1%})")
print(f"  Valid prices: {df['implicit_price'].notna().sum():,} ({df['implicit_price'].notna().mean():.1%})")

# Add transaction size categories
df['txn_size_category'] = 'Unknown'
valid_sol = df['sol_amount'] > 0
df.loc[valid_sol & (df['sol_amount'] >= 100), 'txn_size_category'] = 'Whale'
df.loc[valid_sol & (df['sol_amount'] >= 10) & (df['sol_amount'] < 100), 'txn_size_category'] = 'Big'
df.loc[valid_sol & (df['sol_amount'] >= 1) & (df['sol_amount'] < 10), 'txn_size_category'] = 'Medium'
df.loc[valid_sol & (df['sol_amount'] > 0) & (df['sol_amount'] < 1), 'txn_size_category'] = 'Small'

print(f"\nTransaction size distribution:")
print(df['txn_size_category'].value_counts())

# Show sample of enhanced data
print(f"\n=== ENHANCED DATA SAMPLE ===")
sample_cols = ['coin_name', 'block_timestamp', 'is_buy', 'sol_amount', 'token_amount', 'implicit_price', 'txn_size_category']
print(df[sample_cols].head(10))
```

```python
# Block-Level Grouping: Group transactions by timestamps to approximate blocks
print("=== BLOCK-LEVEL ANALYSIS SETUP ===")

def create_block_groups(coin_data, time_window_ms=100):
    """
    Group transactions into 'blocks' based on timestamp proximity
    
    Args:
        coin_data: DataFrame with transactions for a single coin
        time_window_ms: Time window in milliseconds to group transactions
    
    Returns:
        DataFrame with block_id added
    """
    
    coin_data = coin_data.sort_values('block_timestamp').copy()
    
    # Convert to milliseconds for grouping
    coin_data['timestamp_ms'] = coin_data['block_timestamp'].astype('int64') // 1_000_000
    
    # Group by time windows
    coin_data['time_group'] = coin_data['timestamp_ms'] // time_window_ms
    
    # Create unique block IDs
    coin_data['block_id'] = coin_data.groupby('time_group').ngroup()
    
    return coin_data

# Apply block grouping to each coin
print("Creating block groups for all coins...")
block_data_list = []

for coin_name in df['coin_name'].unique():
    coin_subset = df[df['coin_name'] == coin_name].copy()
    coin_with_blocks = create_block_groups(coin_subset, time_window_ms=100)  # 100ms window
    block_data_list.append(coin_with_blocks)

# Combine all coins back together
df_blocks = pd.concat(block_data_list, ignore_index=True)

# Block statistics
print(f"\n=== BLOCK GROUPING RESULTS ===")
total_blocks = df_blocks['block_id'].nunique()
avg_txns_per_block = len(df_blocks) / total_blocks

print(f"Total blocks identified: {total_blocks:,}")
print(f"Average transactions per block: {avg_txns_per_block:.2f}")

# Block size distribution
block_sizes = df_blocks.groupby('block_id').size()
print(f"\nBlock size distribution:")
print(f"  Single transaction blocks: {(block_sizes == 1).sum():,} ({(block_sizes == 1).mean():.1%})")
print(f"  Multi-transaction blocks: {(block_sizes > 1).sum():,} ({(block_sizes > 1).mean():.1%})")
print(f"  Largest block: {block_sizes.max()} transactions")

# Show block size percentiles
print(f"\nBlock size percentiles:")
for pct in [50, 75, 90, 95, 99]:
    print(f"  {pct}th percentile: {block_sizes.quantile(pct/100):.1f} transactions")

# Sample blocks
print(f"\n=== SAMPLE BLOCKS ===")
large_blocks = block_sizes[block_sizes >= 5].head(3)
for block_id in large_blocks.index:
    block_txns = df_blocks[df_blocks['block_id'] == block_id]
    print(f"\nBlock {block_id} ({len(block_txns)} transactions):")
    print(f"  Coin: {block_txns['coin_name'].iloc[0]}")
    print(f"  Time: {block_txns['block_timestamp'].min()} to {block_txns['block_timestamp'].max()}")
    print(f"  Buys: {block_txns['is_buy'].sum()}, Sells: {block_txns['is_sell'].sum()}")
    print(f"  SOL volume: {block_txns['sol_amount'].sum():.2f}")
    if block_txns['implicit_price'].notna().any():
        prices = block_txns['implicit_price'].dropna()
        print(f"  Price range: {prices.min():.8f} to {prices.max():.8f} SOL/token")
```

```python
# 1. BLOCK-LEVEL ORDER FLOW ANALYSIS
print("=== 1. BLOCK-LEVEL ORDER FLOW PATTERNS ===")

def analyze_block_order_flow(df_blocks, min_block_size=2):
    """
    Analyze order flow patterns within blocks
    """
    
    # Focus on multi-transaction blocks for meaningful analysis
    multi_txn_blocks = df_blocks.groupby('block_id').size()
    multi_block_ids = multi_txn_blocks[multi_txn_blocks >= min_block_size].index
    
    block_analysis = []
    
    for block_id in multi_block_ids:
        block_data = df_blocks[df_blocks['block_id'] == block_id].sort_values('block_timestamp')
        
        # Skip blocks without valid SOL amounts
        if block_data['sol_amount'].sum() == 0:
            continue
            
        analysis = {
            'block_id': block_id,
            'coin_name': block_data['coin_name'].iloc[0],
            'block_size': len(block_data),
            'duration_ms': (block_data['block_timestamp'].max() - block_data['block_timestamp'].min()).total_seconds() * 1000,
            
            # Order flow metrics
            'total_volume': block_data['sol_amount'].sum(),
            'buy_count': block_data['is_buy'].sum(),
            'sell_count': block_data['is_sell'].sum(),
            'buy_volume': block_data[block_data['is_buy']]['sol_amount'].sum(),
            'sell_volume': block_data[block_data['is_sell']]['sol_amount'].sum(),
            
            # Flow imbalance
            'flow_imbalance': (block_data['is_buy'].sum() - block_data['is_sell'].sum()) / len(block_data),
            'volume_imbalance': (block_data[block_data['is_buy']]['sol_amount'].sum() - 
                               block_data[block_data['is_sell']]['sol_amount'].sum()) / block_data['sol_amount'].sum(),
            
            # Transaction sequence patterns
            'first_txn_is_buy': block_data['is_buy'].iloc[0],
            'last_txn_is_buy': block_data['is_buy'].iloc[-1],
            'dominant_direction': 'buy' if block_data['is_buy'].sum() > block_data['is_sell'].sum() else 'sell',
            
            # Size distribution
            'avg_txn_size': block_data['sol_amount'].mean(),
            'largest_txn': block_data['sol_amount'].max(),
            'size_concentration': block_data['sol_amount'].max() / block_data['sol_amount'].sum(),
        }
        
        # Price analysis if available
        if block_data['implicit_price'].notna().any():
            prices = block_data['implicit_price'].dropna()
            analysis.update({
                'price_start': prices.iloc[0] if len(prices) > 0 else np.nan,
                'price_end': prices.iloc[-1] if len(prices) > 0 else np.nan,
                'price_min': prices.min(),
                'price_max': prices.max(),
                'price_volatility': prices.std() / prices.mean() if len(prices) > 1 and prices.mean() > 0 else 0,
                'price_change': (prices.iloc[-1] - prices.iloc[0]) / prices.iloc[0] if len(prices) > 1 and prices.iloc[0] > 0 else 0
            })
        
        block_analysis.append(analysis)
    
    return pd.DataFrame(block_analysis)

# Perform block-level analysis
print("Analyzing order flow patterns within blocks...")
block_flow_df = analyze_block_order_flow(df_blocks, min_block_size=2)

print(f"\nAnalyzed {len(block_flow_df)} multi-transaction blocks")
print(f"Block size range: {block_flow_df['block_size'].min()} to {block_flow_df['block_size'].max()} transactions")
print(f"Volume range: {block_flow_df['total_volume'].min():.2f} to {block_flow_df['total_volume'].max():.2f} SOL")

# Block flow imbalance distribution
print(f"\n=== BLOCK FLOW IMBALANCE PATTERNS ===")
print(f"Flow imbalance distribution:")
print(f"  Strong buy bias (>0.5): {(block_flow_df['flow_imbalance'] > 0.5).sum()} blocks")
print(f"  Moderate buy bias (0.1-0.5): {((block_flow_df['flow_imbalance'] > 0.1) & (block_flow_df['flow_imbalance'] <= 0.5)).sum()} blocks")
print(f"  Balanced (-0.1 to 0.1): {((block_flow_df['flow_imbalance'] >= -0.1) & (block_flow_df['flow_imbalance'] <= 0.1)).sum()} blocks")
print(f"  Moderate sell bias (-0.5 to -0.1): {((block_flow_df['flow_imbalance'] >= -0.5) & (block_flow_df['flow_imbalance'] < -0.1)).sum()} blocks")
print(f"  Strong sell bias (<-0.5): {(block_flow_df['flow_imbalance'] < -0.5).sum()} blocks")

# Visualize block patterns
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Block size distribution
axes[0,0].hist(block_flow_df['block_size'], bins=20, alpha=0.7, edgecolor='black')
axes[0,0].set_title('Block Size Distribution')
axes[0,0].set_xlabel('Number of Transactions per Block')
axes[0,0].set_ylabel('Number of Blocks')

# 2. Flow imbalance distribution
axes[0,1].hist(block_flow_df['flow_imbalance'], bins=30, alpha=0.7, edgecolor='black')
axes[0,1].axvline(x=0, color='red', linestyle='--', alpha=0.7, label='Balanced')
axes[0,1].set_title('Block Flow Imbalance Distribution')
axes[0,1].set_xlabel('Flow Imbalance (Buy - Sell) / Total')
axes[0,1].set_ylabel('Number of Blocks')
axes[0,1].legend()

# 3. Volume vs Flow Imbalance
scatter = axes[1,0].scatter(block_flow_df['total_volume'], block_flow_df['flow_imbalance'], 
                           c=block_flow_df['block_size'], cmap='viridis', alpha=0.6)
axes[1,0].set_xlabel('Total Block Volume (SOL)')
axes[1,0].set_ylabel('Flow Imbalance')
axes[1,0].set_title('Volume vs Flow Imbalance (colored by block size)')
plt.colorbar(scatter, ax=axes[1,0], label='Block Size')

# 4. Block duration vs size
if 'duration_ms' in block_flow_df.columns:
    axes[1,1].scatter(block_flow_df['block_size'], block_flow_df['duration_ms'], alpha=0.6)
    axes[1,1].set_xlabel('Block Size (transactions)')
    axes[1,1].set_ylabel('Block Duration (ms)')
    axes[1,1].set_title('Block Size vs Duration')
else:
    axes[1,1].text(0.5, 0.5, 'Duration data not available', ha='center', va='center', transform=axes[1,1].transAxes)

plt.tight_layout()
plt.show()
```

```python
# 1.2 TRANSACTION SEQUENCING PATTERNS
print("=== 1.2 TRANSACTION SEQUENCING WITHIN BLOCKS ===")

def analyze_transaction_sequences(df_blocks, min_block_size=3):
    """
    Analyze the sequence patterns of transactions within blocks
    """
    
    # Get blocks with enough transactions for sequence analysis
    block_sizes = df_blocks.groupby('block_id').size()
    sequence_blocks = block_sizes[block_sizes >= min_block_size].index
    
    sequence_patterns = {
        'buy_sell_patterns': [],
        'size_patterns': [],
        'momentum_patterns': []
    }
    
    for block_id in sequence_blocks:
        block_data = df_blocks[df_blocks['block_id'] == block_id].sort_values('block_timestamp')
        
        if block_data['sol_amount'].sum() == 0:
            continue
            
        # Analyze buy/sell sequence
        directions = ['B' if is_buy else 'S' for is_buy in block_data['is_buy']]
        direction_sequence = ''.join(directions)
        
        # Analyze size patterns (relative to block average)
        avg_size = block_data['sol_amount'].mean()
        size_categories = ['L' if size > avg_size else 'S' for size in block_data['sol_amount']]
        size_sequence = ''.join(size_categories)
        
        # Analyze momentum (price change patterns if available)
        momentum_sequence = ''
        if block_data['implicit_price'].notna().sum() >= 2:
            prices = block_data['implicit_price'].dropna()
            if len(prices) >= 2:
                price_changes = prices.diff().dropna()
                momentum_sequence = ''.join(['U' if change > 0 else 'D' if change < 0 else 'F' for change in price_changes])
        
        sequence_patterns['buy_sell_patterns'].append(direction_sequence)
        sequence_patterns['size_patterns'].append(size_sequence)
        if momentum_sequence:
            sequence_patterns['momentum_patterns'].append(momentum_sequence)
    
    return sequence_patterns

# Analyze sequences
print("Analyzing transaction sequences within blocks...")
sequences = analyze_transaction_sequences(df_blocks, min_block_size=3)

# Analyze buy/sell patterns
print(f"\n=== BUY/SELL SEQUENCE PATTERNS ===")
print(f"Analyzed {len(sequences['buy_sell_patterns'])} blocks with 3+ transactions")

# Count common patterns
from collections import Counter
pattern_counts = Counter(sequences['buy_sell_patterns'])
most_common_patterns = pattern_counts.most_common(10)

print(f"\nMost common buy/sell patterns:")
for pattern, count in most_common_patterns:
    pct = count / len(sequences['buy_sell_patterns']) * 100
    print(f"  {pattern}: {count} blocks ({pct:.1f}%)")

# Analyze specific patterns of interest
def analyze_specific_patterns(patterns):
    """Analyze specific types of patterns"""
    results = {
        'alternating': 0,  # BSBSBS...
        'clustered_buys': 0,  # BBB...SSS
        'clustered_sells': 0,  # SSS...BBB
        'sandwich_buy': 0,  # Buy-Sell-Buy
        'sandwich_sell': 0,  # Sell-Buy-Sell
        'momentum_buy': 0,  # Ending with multiple buys
        'momentum_sell': 0,  # Ending with multiple sells
    }
    
    for pattern in patterns:
        # Alternating pattern
        is_alternating = True
        for i in range(1, len(pattern)):
            if pattern[i] == pattern[i-1]:
                is_alternating = False
                break
        if is_alternating and len(pattern) >= 4:
            results['alternating'] += 1
        
        # Clustered patterns
        if 'BBB' in pattern:
            results['clustered_buys'] += 1
        if 'SSS' in pattern:
            results['clustered_sells'] += 1
            
        # Sandwich patterns
        if 'BSB' in pattern:
            results['sandwich_buy'] += 1
        if 'SBS' in pattern:
            results['sandwich_sell'] += 1
            
        # Momentum patterns (ending with 2+ same direction)
        if pattern.endswith('BB') or pattern.endswith('BBB'):
            results['momentum_buy'] += 1
        if pattern.endswith('SS') or pattern.endswith('SSS'):
            results['momentum_sell'] += 1
    
    return results

pattern_analysis = analyze_specific_patterns(sequences['buy_sell_patterns'])

print(f"\n=== SPECIAL PATTERN ANALYSIS ===")
total_patterns = len(sequences['buy_sell_patterns'])
for pattern_type, count in pattern_analysis.items():
    pct = count / total_patterns * 100
    print(f"  {pattern_type.replace('_', ' ').title()}: {count} blocks ({pct:.1f}%)")

# Visualize pattern distributions
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# 1. Pattern length distribution
pattern_lengths = [len(p) for p in sequences['buy_sell_patterns']]
axes[0].hist(pattern_lengths, bins=range(3, max(pattern_lengths)+2), alpha=0.7, edgecolor='black')
axes[0].set_title('Transaction Sequence Length Distribution')
axes[0].set_xlabel('Number of Transactions in Block')
axes[0].set_ylabel('Number of Blocks')

# 2. Buy/Sell ratio in patterns
buy_ratios = [p.count('B') / len(p) for p in sequences['buy_sell_patterns']]
axes[1].hist(buy_ratios, bins=20, alpha=0.7, edgecolor='black')
axes[1].axvline(x=0.5, color='red', linestyle='--', alpha=0.7, label='50/50 Split')
axes[1].set_title('Buy Ratio Distribution in Sequences')
axes[1].set_xlabel('Proportion of Buys in Block')
axes[1].set_ylabel('Number of Blocks')
axes[1].legend()

# 3. Special patterns frequency
pattern_names = list(pattern_analysis.keys())
pattern_counts = list(pattern_analysis.values())
axes[2].bar(range(len(pattern_names)), pattern_counts, alpha=0.7)
axes[2].set_title('Special Pattern Frequencies')
axes[2].set_xlabel('Pattern Type')
axes[2].set_ylabel('Number of Blocks')
axes[2].set_xticks(range(len(pattern_names)))
axes[2].set_xticklabels([name.replace('_', '\n').title() for name in pattern_names], rotation=45)

plt.tight_layout()
plt.show()

# Show some interesting sequence examples
print(f"\n=== INTERESTING SEQUENCE EXAMPLES ===")
interesting_patterns = ['BSBSBS', 'BBSSBB', 'SSBSSS', 'BBBSSS', 'SSSBBB']
for pattern in interesting_patterns:
    if pattern in pattern_counts:
        print(f"Pattern '{pattern}': {pattern_counts[pattern]} occurrences")
        # Find an example
        for i, p in enumerate(sequences['buy_sell_patterns']):
            if p == pattern:
                print(f"  Example: Block with {len(pattern)} transactions showing {pattern} pattern")
                break
```

```python
# 2. FLOW TOXICITY ANALYSIS
print("=== 2. FLOW TOXICITY & INFORMATION CONTENT EDA ===")

def calculate_flow_toxicity_metrics(df_blocks):
    """
    Calculate flow toxicity metrics based on Kyle's lambda and information content
    Using SOL-relative pricing for analysis
    """
    
    # Focus on coins with sufficient price data
    coin_toxicity = []
    
    for coin_name in df_blocks['coin_name'].unique():
        coin_data = df_blocks[df_blocks['coin_name'] == coin_name].copy()
        
        # Skip coins without sufficient price data
        valid_prices = coin_data['implicit_price'].notna().sum()
        if valid_prices < 20:  # Need at least 20 price observations
            continue
            
        coin_data = coin_data[coin_data['implicit_price'].notna()].sort_values('block_timestamp')
        
        # Calculate returns
        coin_data['price_return'] = coin_data['implicit_price'].pct_change()
        coin_data['price_change'] = coin_data['implicit_price'].diff()
        
        # Calculate order flow imbalance (signed volume)
        coin_data['signed_volume'] = np.where(coin_data['is_buy'], 
                                            coin_data['sol_amount'], 
                                            -coin_data['sol_amount'])
        
        # Rolling metrics for toxicity analysis
        window_size = min(10, len(coin_data) // 4)  # Adaptive window size
        if window_size < 5:
            continue
            
        # Kyle's Lambda approximation: price_impact / signed_volume
        # Using rolling correlation as proxy for lambda
        coin_data['rolling_price_impact'] = coin_data['price_change'].rolling(window_size).std()
        coin_data['rolling_volume'] = coin_data['signed_volume'].abs().rolling(window_size).mean()
        
        # Information content metrics
        coin_data['return_volatility'] = coin_data['price_return'].rolling(window_size).std()
        coin_data['volume_volatility'] = coin_data['sol_amount'].rolling(window_size).std()
        
        # Flow-price correlation (toxicity indicator)
        def rolling_correlation(series1, series2, window):
            return series1.rolling(window).corr(series2)
        
        coin_data['flow_price_corr'] = rolling_correlation(
            coin_data['signed_volume'], 
            coin_data['price_change'], 
            window_size
        )
        
        # Aggregate metrics
        analysis = {
            'coin_name': coin_name,
            'total_transactions': len(coin_data),
            'price_observations': valid_prices,
            'time_span_hours': (coin_data['block_timestamp'].max() - coin_data['block_timestamp'].min()).total_seconds() / 3600,
            
            # Price metrics
            'avg_price': coin_data['implicit_price'].mean(),
            'price_volatility': coin_data['price_return'].std(),
            'price_range_pct': (coin_data['implicit_price'].max() - coin_data['implicit_price'].min()) / coin_data['implicit_price'].mean(),
            
            # Volume metrics
            'total_volume': coin_data['sol_amount'].sum(),
            'avg_trade_size': coin_data['sol_amount'].mean(),
            'volume_concentration': coin_data['sol_amount'].max() / coin_data['sol_amount'].sum(),
            
            # Flow imbalance
            'net_flow': coin_data['signed_volume'].sum(),
            'flow_imbalance': coin_data['signed_volume'].sum() / coin_data['sol_amount'].sum(),
            'buy_ratio': coin_data['is_buy'].mean(),
            
            # Toxicity proxies
            'avg_flow_price_corr': coin_data['flow_price_corr'].mean(),
            'flow_price_corr_volatility': coin_data['flow_price_corr'].std(),
            'kyle_lambda_proxy': (coin_data['rolling_price_impact'] / coin_data['rolling_volume']).mean(),
            
            # Information content
            'price_volume_corr': coin_data['implicit_price'].corr(coin_data['sol_amount']),
            'return_volume_corr': coin_data['price_return'].corr(coin_data['sol_amount']),
            'informed_trading_proxy': abs(coin_data['flow_price_corr'].mean()),
        }
        
        coin_toxicity.append(analysis)
    
    return pd.DataFrame(coin_toxicity)

# Calculate toxicity metrics
print("Calculating flow toxicity metrics for each coin...")
toxicity_df = calculate_flow_toxicity_metrics(df_blocks)

print(f"\nAnalyzed {len(toxicity_df)} coins with sufficient price data")
print(f"Coins analyzed: {', '.join(toxicity_df['coin_name'].head(10).tolist())}")

# Toxicity distribution analysis
print(f"\n=== FLOW TOXICITY DISTRIBUTION ===")

# Define toxicity categories based on flow-price correlation
toxicity_df['toxicity_category'] = 'Unknown'
flow_corr_col = 'avg_flow_price_corr'

# Handle NaN values
valid_corr = toxicity_df[flow_corr_col].notna()
if valid_corr.sum() > 0:
    q75 = toxicity_df.loc[valid_corr, flow_corr_col].quantile(0.75)
    q25 = toxicity_df.loc[valid_corr, flow_corr_col].quantile(0.25)
    
    toxicity_df.loc[valid_corr & (toxicity_df[flow_corr_col] > q75), 'toxicity_category'] = 'High Toxicity'
    toxicity_df.loc[valid_corr & (toxicity_df[flow_corr_col] > q25) & (toxicity_df[flow_corr_col] <= q75), 'toxicity_category'] = 'Medium Toxicity'
    toxicity_df.loc[valid_corr & (toxicity_df[flow_corr_col] <= q25), 'toxicity_category'] = 'Low Toxicity'

print(f"Toxicity category distribution:")
print(toxicity_df['toxicity_category'].value_counts())

# Display top toxic and least toxic coins
if valid_corr.sum() > 0:
    print(f"\n=== MOST TOXIC FLOW (Highest Information Content) ===")
    most_toxic = toxicity_df.nlargest(5, flow_corr_col)
    for _, row in most_toxic.iterrows():
        print(f"  {row['coin_name']}: Flow-Price Corr = {row[flow_corr_col]:.3f}, Price Vol = {row['price_volatility']:.3f}")
    
    print(f"\n=== LEAST TOXIC FLOW (Lowest Information Content) ===")
    least_toxic = toxicity_df.nsmallest(5, flow_corr_col)
    for _, row in least_toxic.iterrows():
        print(f"  {row['coin_name']}: Flow-Price Corr = {row[flow_corr_col]:.3f}, Price Vol = {row['price_volatility']:.3f}")

# Visualize toxicity patterns
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Flow-Price Correlation Distribution
if valid_corr.sum() > 0:
    axes[0,0].hist(toxicity_df[flow_corr_col].dropna(), bins=20, alpha=0.7, edgecolor='black')
    axes[0,0].axvline(x=0, color='red', linestyle='--', alpha=0.7, label='Zero Correlation')
    axes[0,0].set_title('Flow-Price Correlation Distribution')
    axes[0,0].set_xlabel('Average Flow-Price Correlation')
    axes[0,0].set_ylabel('Number of Coins')
    axes[0,0].legend()
else:
    axes[0,0].text(0.5, 0.5, 'Insufficient correlation data', ha='center', va='center', transform=axes[0,0].transAxes)

# 2. Price Volatility vs Flow Correlation
if valid_corr.sum() > 0:
    scatter = axes[0,1].scatter(toxicity_df['price_volatility'], toxicity_df[flow_corr_col], 
                              c=toxicity_df['total_volume'], cmap='viridis', alpha=0.6)
    axes[0,1].set_xlabel('Price Volatility')
    axes[0,1].set_ylabel('Flow-Price Correlation')
    axes[0,1].set_title('Volatility vs Flow Toxicity (colored by volume)')
    plt.colorbar(scatter, ax=axes[0,1], label='Total Volume')
else:
    axes[0,1].text(0.5, 0.5, 'Insufficient data for scatter plot', ha='center', va='center', transform=axes[0,1].transAxes)

# 3. Volume Concentration vs Toxicity
if valid_corr.sum() > 0:
    axes[1,0].scatter(toxicity_df['volume_concentration'], toxicity_df[flow_corr_col], alpha=0.6)
    axes[1,0].set_xlabel('Volume Concentration (Max Trade / Total Volume)')
    axes[1,0].set_ylabel('Flow-Price Correlation')
    axes[1,0].set_title('Volume Concentration vs Flow Toxicity')
else:
    axes[1,0].text(0.5, 0.5, 'Insufficient data for scatter plot', ha='center', va='center', transform=axes[1,0].transAxes)

# 4. Buy Ratio vs Toxicity
if valid_corr.sum() > 0:
    axes[1,1].scatter(toxicity_df['buy_ratio'], toxicity_df[flow_corr_col], alpha=0.6)
    axes[1,1].set_xlabel('Buy Ratio')
    axes[1,1].set_ylabel('Flow-Price Correlation')
    axes[1,1].set_title('Buy Ratio vs Flow Toxicity')
else:
    axes[1,1].text(0.5, 0.5, 'Insufficient data for scatter plot', ha='center', va='center', transform=axes[1,1].transAxes)

plt.tight_layout()
plt.show()
```

```python
# 3. INTEGRATED ANALYSIS: Block Patterns + Flow Toxicity
print("=== 3. INTEGRATED BLOCK-LEVEL & TOXICITY ANALYSIS ===")

def integrate_block_toxicity_analysis(block_flow_df, toxicity_df):
    """
    Integrate block-level patterns with flow toxicity metrics
    """
    
    # Merge datasets on coin_name
    integrated_df = block_flow_df.merge(
        toxicity_df[['coin_name', 'avg_flow_price_corr', 'price_volatility', 
                    'toxicity_category', 'informed_trading_proxy']], 
        on='coin_name', 
        how='inner'
    )
    
    return integrated_df

# Create integrated analysis
print("Integrating block patterns with toxicity metrics...")
if len(toxicity_df) > 0 and len(block_flow_df) > 0:
    integrated_df = integrate_block_toxicity_analysis(block_flow_df, toxicity_df)
    
    print(f"Integrated analysis covers {len(integrated_df)} blocks from {integrated_df['coin_name'].nunique()} coins")
    
    # Analyze relationships
    print(f"\n=== BLOCK PATTERNS BY TOXICITY CATEGORY ===")
    
    if 'toxicity_category' in integrated_df.columns:
        for category in integrated_df['toxicity_category'].unique():
            if category == 'Unknown':
                continue
                
            category_data = integrated_df[integrated_df['toxicity_category'] == category]
            
            print(f"\n{category} Coins ({len(category_data)} blocks):")
            print(f"  Average block size: {category_data['block_size'].mean():.1f} transactions")
            print(f"  Average flow imbalance: {category_data['flow_imbalance'].mean():.3f}")
            print(f"  Average volume imbalance: {category_data['volume_imbalance'].mean():.3f}")
            print(f"  Dominant buy blocks: {(category_data['dominant_direction'] == 'buy').mean():.1%}")
            print(f"  Average block volume: {category_data['total_volume'].mean():.2f} SOL")
    
    # Statistical tests for toxicity differences
    if integrated_df['toxicity_category'].nunique() > 1:
        print(f"\n=== STATISTICAL ANALYSIS ===")
        
        # Test if high toxicity coins have different block patterns
        high_toxicity = integrated_df[integrated_df['toxicity_category'] == 'High Toxicity']
        low_toxicity = integrated_df[integrated_df['toxicity_category'] == 'Low Toxicity']
        
        if len(high_toxicity) > 0 and len(low_toxicity) > 0:
            # T-test for flow imbalance
            from scipy.stats import ttest_ind
            t_stat, p_val = ttest_ind(high_toxicity['flow_imbalance'], low_toxicity['flow_imbalance'])
            print(f"Flow imbalance difference (High vs Low toxicity):")
            print(f"  High toxicity mean: {high_toxicity['flow_imbalance'].mean():.3f}")
            print(f"  Low toxicity mean: {low_toxicity['flow_imbalance'].mean():.3f}")
            print(f"  T-statistic: {t_stat:.3f}, P-value: {p_val:.3f}")
            
            # T-test for block size
            t_stat, p_val = ttest_ind(high_toxicity['block_size'], low_toxicity['block_size'])
            print(f"\nBlock size difference (High vs Low toxicity):")
            print(f"  High toxicity mean: {high_toxicity['block_size'].mean():.1f}")
            print(f"  Low toxicity mean: {low_toxicity['block_size'].mean():.1f}")
            print(f"  T-statistic: {t_stat:.3f}, P-value: {p_val:.3f}")
    
    # Visualization of integrated patterns
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # 1. Block size vs toxicity
    if 'avg_flow_price_corr' in integrated_df.columns:
        scatter = axes[0,0].scatter(integrated_df['block_size'], integrated_df['avg_flow_price_corr'], 
                                   c=integrated_df['total_volume'], cmap='plasma', alpha=0.6)
        axes[0,0].set_xlabel('Block Size (transactions)')
        axes[0,0].set_ylabel('Flow-Price Correlation (Toxicity)')
        axes[0,0].set_title('Block Size vs Flow Toxicity')
        plt.colorbar(scatter, ax=axes[0,0], label='Block Volume (SOL)')
    
    # 2. Flow imbalance vs toxicity
    if 'avg_flow_price_corr' in integrated_df.columns:
        axes[0,1].scatter(integrated_df['flow_imbalance'], integrated_df['avg_flow_price_corr'], alpha=0.6)
        axes[0,1].set_xlabel('Block Flow Imbalance')
        axes[0,1].set_ylabel('Flow-Price Correlation (Toxicity)')
        axes[0,1].set_title('Flow Imbalance vs Toxicity')
    
    # 3. Toxicity by dominant direction
    if 'toxicity_category' in integrated_df.columns:
        toxicity_by_direction = integrated_df.groupby(['dominant_direction', 'toxicity_category']).size().unstack(fill_value=0)
        if not toxicity_by_direction.empty:
            toxicity_by_direction.plot(kind='bar', ax=axes[1,0], alpha=0.7)
            axes[1,0].set_title('Block Direction by Toxicity Category')
            axes[1,0].set_xlabel('Dominant Direction')
            axes[1,0].set_ylabel('Number of Blocks')
            axes[1,0].legend(title='Toxicity Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 4. Volume concentration patterns
    if 'avg_flow_price_corr' in integrated_df.columns:
        axes[1,1].scatter(integrated_df['size_concentration'], integrated_df['avg_flow_price_corr'], alpha=0.6)
        axes[1,1].set_xlabel('Size Concentration (Largest Trade / Block Volume)')
        axes[1,1].set_ylabel('Flow-Price Correlation (Toxicity)')
        axes[1,1].set_title('Size Concentration vs Toxicity')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("Insufficient data for integrated analysis")

# Summary insights
print(f"\n=== SUMMARY INSIGHTS ===")
print("Key discoveries from block-level order flow and toxicity analysis:")
print()

if len(block_flow_df) > 0:
    print("Block-Level Patterns:")
    print(f"  • Analyzed {len(block_flow_df)} multi-transaction blocks")
    print(f"  • Average block size: {block_flow_df['block_size'].mean():.1f} transactions")
    print(f"  • Flow imbalanced blocks (|imbalance| > 0.3): {(abs(block_flow_df['flow_imbalance']) > 0.3).mean():.1%}")
    
    if len(sequences['buy_sell_patterns']) > 0:
        # Calculate most common sequence length
        pattern_lengths = [len(p) for p in sequences['buy_sell_patterns']]
        length_counts = Counter(pattern_lengths)
        most_common_length = length_counts.most_common(1)[0][0]
        print(f"  • Most common sequence length: {most_common_length} transactions")
        print(f"  • Alternating patterns: {pattern_analysis.get('alternating', 0)} blocks")
        print(f"  • Momentum patterns: {pattern_analysis.get('momentum_buy', 0) + pattern_analysis.get('momentum_sell', 0)} blocks")

if len(toxicity_df) > 0:
    print(f"\nFlow Toxicity Patterns:")
    print(f"  • Analyzed {len(toxicity_df)} coins with sufficient price data")
    valid_corr = toxicity_df['avg_flow_price_corr'].notna()
    if valid_corr.sum() > 0:
        print(f"  • Average flow-price correlation: {toxicity_df.loc[valid_corr, 'avg_flow_price_corr'].mean():.3f}")
        print(f"  • High toxicity coins: {(toxicity_df['toxicity_category'] == 'High Toxicity').sum()}")
        print(f"  • Price volatility range: {toxicity_df['price_volatility'].min():.3f} to {toxicity_df['price_volatility'].max():.3f}")

if 'integrated_df' in locals() and len(integrated_df) > 0:
    print(f"\nIntegrated Findings:")
    print(f"  • Blocks analyzed across both dimensions: {len(integrated_df)}")
    print(f"  • Coins with both block and toxicity data: {integrated_df['coin_name'].nunique()}")
    
print(f"\nNext steps for deeper analysis:")
print(f"  • Temporal analysis of toxicity changes")
print(f"  • Cross-coin contamination effects") 
print(f"  • Regime detection in flow patterns")
print(f"  • Predictive modeling opportunities")
```

```python
# ENHANCED ANALYSIS: Transaction Sequencing with Trader Factor
print("=== ENHANCED TRANSACTION SEQUENCING - ACCOUNTING FOR TRADERS ===")

def analyze_trader_aware_sequences(df_blocks, min_block_size=3):
    """
    Analyze transaction sequences accounting for individual traders
    """
    
    block_sizes = df_blocks.groupby('block_id').size()
    sequence_blocks = block_sizes[block_sizes >= min_block_size].index
    
    trader_analysis = {
        'single_trader_blocks': 0,
        'multi_trader_blocks': 0,
        'single_trader_patterns': [],
        'multi_trader_patterns': [],
        'trader_sequence_patterns': [],
        'same_trader_consecutive': 0,
        'trader_alternating': 0
    }
    
    for block_id in sequence_blocks:
        block_data = df_blocks[df_blocks['block_id'] == block_id].sort_values('block_timestamp')
        
        if block_data['sol_amount'].sum() == 0:
            continue
        
        # Check if block has single or multiple traders
        unique_traders = block_data['swapper'].nunique()
        
        # Analyze buy/sell sequence
        directions = ['B' if is_buy else 'S' for is_buy in block_data['is_buy']]
        direction_sequence = ''.join(directions)
        
        # Analyze trader sequence
        traders = block_data['swapper'].tolist()
        trader_changes = sum(1 for i in range(1, len(traders)) if traders[i] != traders[i-1])
        
        if unique_traders == 1:
            trader_analysis['single_trader_blocks'] += 1
            trader_analysis['single_trader_patterns'].append(direction_sequence)
        else:
            trader_analysis['multi_trader_blocks'] += 1
            trader_analysis['multi_trader_patterns'].append(direction_sequence)
            
            # Analyze trader-direction patterns
            trader_direction_pattern = []
            for i, (trader, direction) in enumerate(zip(traders, directions)):
                if i == 0:
                    trader_direction_pattern.append(f"T1{direction}")
                else:
                    if trader == traders[i-1]:
                        trader_direction_pattern.append(f"Same{direction}")
                    else:
                        trader_direction_pattern.append(f"New{direction}")
            
            trader_analysis['trader_sequence_patterns'].append(''.join(trader_direction_pattern))
        
        # Count consecutive same-trader transactions
        consecutive_same = 0
        for i in range(1, len(traders)):
            if traders[i] == traders[i-1]:
                consecutive_same += 1
        trader_analysis['same_trader_consecutive'] += consecutive_same
        
        # Check for trader alternating pattern
        if len(set(traders)) == 2 and trader_changes >= len(traders) - 2:
            trader_analysis['trader_alternating'] += 1
    
    return trader_analysis

# Perform trader-aware analysis
print("Analyzing transaction sequences with trader identification...")
trader_sequences = analyze_trader_aware_sequences(df_blocks, min_block_size=3)

print(f"\n=== TRADER-AWARE SEQUENCE ANALYSIS ===")
total_analyzed = trader_sequences['single_trader_blocks'] + trader_sequences['multi_trader_blocks']
print(f"Total blocks analyzed: {total_analyzed}")
print(f"Single-trader blocks: {trader_sequences['single_trader_blocks']} ({trader_sequences['single_trader_blocks']/total_analyzed:.1%})")
print(f"Multi-trader blocks: {trader_sequences['multi_trader_blocks']} ({trader_sequences['multi_trader_blocks']/total_analyzed:.1%})")

# Analyze single-trader patterns vs multi-trader patterns
print(f"\n=== SINGLE-TRADER PATTERN ANALYSIS ===")
if trader_sequences['single_trader_patterns']:
    single_trader_counts = Counter(trader_sequences['single_trader_patterns'])
    print(f"Most common single-trader patterns:")
    for pattern, count in single_trader_counts.most_common(10):
        pct = count / len(trader_sequences['single_trader_patterns']) * 100
        print(f"  {pattern}: {count} blocks ({pct:.1f}%)")

print(f"\n=== MULTI-TRADER PATTERN ANALYSIS ===")
if trader_sequences['multi_trader_patterns']:
    multi_trader_counts = Counter(trader_sequences['multi_trader_patterns'])
    print(f"Most common multi-trader patterns:")
    for pattern, count in multi_trader_counts.most_common(10):
        pct = count / len(trader_sequences['multi_trader_patterns']) * 100
        print(f"  {pattern}: {count} blocks ({pct:.1f}%)")

# Compare pattern distributions
print(f"\n=== PATTERN COMPARISON: Single vs Multi-Trader ===")

def analyze_pattern_characteristics(patterns, label):
    if not patterns:
        print(f"{label}: No patterns to analyze")
        return
    
    # Pattern length distribution
    lengths = [len(p) for p in patterns]
    avg_length = np.mean(lengths)
    
    # Buy ratio distribution
    buy_ratios = [p.count('B') / len(p) for p in patterns]
    avg_buy_ratio = np.mean(buy_ratios)
    
    # Clustering analysis
    clustered_buys = sum(1 for p in patterns if 'BBB' in p)
    clustered_sells = sum(1 for p in patterns if 'SSS' in p)
    alternating = sum(1 for p in patterns if all(p[i] != p[i+1] for i in range(len(p)-1)) and len(p) >= 4)
    
    print(f"{label} ({len(patterns)} blocks):")
    print(f"  Average pattern length: {avg_length:.1f}")
    print(f"  Average buy ratio: {avg_buy_ratio:.1%}")
    print(f"  Clustered buys (BBB+): {clustered_buys} ({clustered_buys/len(patterns):.1%})")
    print(f"  Clustered sells (SSS+): {clustered_sells} ({clustered_sells/len(patterns):.1%})")
    print(f"  Alternating patterns: {alternating} ({alternating/len(patterns):.1%})")

analyze_pattern_characteristics(trader_sequences['single_trader_patterns'], "Single-Trader Blocks")
analyze_pattern_characteristics(trader_sequences['multi_trader_patterns'], "Multi-Trader Blocks")

# Visualize the differences
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Block type distribution
block_types = ['Single Trader', 'Multi Trader']
block_counts = [trader_sequences['single_trader_blocks'], trader_sequences['multi_trader_blocks']]
axes[0,0].pie(block_counts, labels=block_types, autopct='%1.1f%%', startangle=90)
axes[0,0].set_title('Block Composition by Trader Count')

# 2. Pattern length comparison
if trader_sequences['single_trader_patterns'] and trader_sequences['multi_trader_patterns']:
    single_lengths = [len(p) for p in trader_sequences['single_trader_patterns']]
    multi_lengths = [len(p) for p in trader_sequences['multi_trader_patterns']]
    
    axes[0,1].hist(single_lengths, alpha=0.6, label='Single Trader', bins=20, density=True)
    axes[0,1].hist(multi_lengths, alpha=0.6, label='Multi Trader', bins=20, density=True)
    axes[0,1].set_xlabel('Pattern Length')
    axes[0,1].set_ylabel('Density')
    axes[0,1].set_title('Pattern Length Distribution')
    axes[0,1].legend()

# 3. Buy ratio comparison
if trader_sequences['single_trader_patterns'] and trader_sequences['multi_trader_patterns']:
    single_buy_ratios = [p.count('B') / len(p) for p in trader_sequences['single_trader_patterns']]
    multi_buy_ratios = [p.count('B') / len(p) for p in trader_sequences['multi_trader_patterns']]
    
    axes[1,0].hist(single_buy_ratios, alpha=0.6, label='Single Trader', bins=20, density=True)
    axes[1,0].hist(multi_buy_ratios, alpha=0.6, label='Multi Trader', bins=20, density=True)
    axes[1,0].set_xlabel('Buy Ratio')
    axes[1,0].set_ylabel('Density')
    axes[1,0].set_title('Buy Ratio Distribution')
    axes[1,0].legend()

# 4. Clustering comparison
if trader_sequences['single_trader_patterns'] and trader_sequences['multi_trader_patterns']:
    single_clustered = [1 if 'BBB' in p or 'SSS' in p else 0 for p in trader_sequences['single_trader_patterns']]
    multi_clustered = [1 if 'BBB' in p or 'SSS' in p else 0 for p in trader_sequences['multi_trader_patterns']]
    
    clustering_data = [
        ['Single Trader', 'Clustered', np.mean(single_clustered)],
        ['Single Trader', 'Not Clustered', 1 - np.mean(single_clustered)],
        ['Multi Trader', 'Clustered', np.mean(multi_clustered)],
        ['Multi Trader', 'Not Clustered', 1 - np.mean(multi_clustered)]
    ]
    
    clustering_df = pd.DataFrame(clustering_data, columns=['Trader_Type', 'Pattern_Type', 'Proportion'])
    pivot_df = clustering_df.pivot(index='Trader_Type', columns='Pattern_Type', values='Proportion')
    pivot_df.plot(kind='bar', ax=axes[1,1], alpha=0.7)
    axes[1,1].set_title('Clustering Patterns by Trader Type')
    axes[1,1].set_ylabel('Proportion')
    axes[1,1].legend(title='Pattern Type')

plt.tight_layout()
plt.show()
```

```python
# FIXED LEAD-LAG ANALYSIS: Simple and Correct Implementation
print("=== FIXED LEAD-LAG ANALYSIS ===")

def simple_leadlag_test(df_blocks):
    """
    Simple, correct lead-lag analysis to test Information vs FOMO
    """
    
    results = []
    
    for coin_name in df_blocks['coin_name'].unique():
        coin_data = df_blocks[df_blocks['coin_name'] == coin_name].copy()
        
        # Need sufficient data
        if len(coin_data) < 100:
            continue
            
        # Sort by time and remove NaN prices
        coin_data = coin_data[coin_data['implicit_price'].notna()].sort_values('block_timestamp')
        
        if len(coin_data) < 50:
            continue
        
        # Calculate basic metrics
        coin_data['price_change'] = coin_data['implicit_price'].diff()
        coin_data['signed_volume'] = np.where(coin_data['is_buy'], 
                                            coin_data['sol_amount'], 
                                            -coin_data['sol_amount'])
        
        # Remove first row (NaN from diff)
        coin_data = coin_data.iloc[1:].copy()
        
        if len(coin_data) < 30:
            continue
        
        # Test 1: Does PAST flow predict FUTURE price changes? (Information hypothesis)
        # Take flow at time t, price change at time t+1
        flow_t = coin_data['signed_volume'].iloc[:-1].values  # t=0 to t=n-1
        price_change_t_plus_1 = coin_data['price_change'].iloc[1:].values  # t=1 to t=n
        
        flow_predicts_price = np.corrcoef(flow_t, price_change_t_plus_1)[0, 1]
        
        # Test 2: Does PAST price change predict FUTURE flow? (FOMO hypothesis)  
        # Take price change at time t, flow at time t+1
        price_change_t = coin_data['price_change'].iloc[:-1].values  # t=0 to t=n-1
        flow_t_plus_1 = coin_data['signed_volume'].iloc[1:].values  # t=1 to t=n
        
        price_predicts_flow = np.corrcoef(price_change_t, flow_t_plus_1)[0, 1]
        
        # Test 3: Momentum test - do large price moves trigger volume surges?
        abs_price_change_t = np.abs(coin_data['price_change'].iloc[:-1].values)
        volume_t_plus_1 = coin_data['sol_amount'].iloc[1:].values
        
        momentum_correlation = np.corrcoef(abs_price_change_t, volume_t_plus_1)[0, 1]
        
        # Interpretation
        interpretation = 'Mixed'
        if abs(flow_predicts_price) > abs(price_predicts_flow) * 1.5:
            interpretation = 'Information-driven'
        elif abs(price_predicts_flow) > abs(flow_predicts_price) * 1.5:
            interpretation = 'FOMO-driven'
        
        results.append({
            'coin_name': coin_name,
            'data_points': len(coin_data),
            'flow_predicts_price': flow_predicts_price,
            'price_predicts_flow': price_predicts_flow,
            'momentum_correlation': momentum_correlation,
            'interpretation': interpretation
        })
    
    return pd.DataFrame(results)

# Run the corrected analysis
print("Running corrected lead-lag analysis...")
fixed_results = simple_leadlag_test(df_blocks)

if len(fixed_results) > 0:
    print(f"\n=== CORRECTED RESULTS ===")
    print(f"Analyzed {len(fixed_results)} coins")
    
    # Show results for each coin
    for _, row in fixed_results.iterrows():
        print(f"\n{row['coin_name']} ({row['data_points']} data points):")
        print(f"  Flow predicts price: {row['flow_predicts_price']:.4f}")
        print(f"  Price predicts flow: {row['price_predicts_flow']:.4f}")
        print(f"  Momentum correlation: {row['momentum_correlation']:.4f}")
        print(f"  Interpretation: {row['interpretation']}")
    
    # Summary statistics
    print(f"\n=== SUMMARY STATISTICS ===")
    print(f"Average flow predicts price: {fixed_results['flow_predicts_price'].mean():.4f}")
    print(f"Average price predicts flow: {fixed_results['price_predicts_flow'].mean():.4f}")
    print(f"Average momentum correlation: {fixed_results['momentum_correlation'].mean():.4f}")
    
    # Count interpretations
    print(f"\nInterpretation breakdown:")
    print(fixed_results['interpretation'].value_counts())
    
    # Test your FOMO hypothesis
    print(f"\n=== TESTING FOMO HYPOTHESIS ===")
    
    # Evidence for FOMO: price_predicts_flow > flow_predicts_price
    fomo_evidence = (fixed_results['price_predicts_flow'].abs() > fixed_results['flow_predicts_price'].abs()).sum()
    info_evidence = (fixed_results['flow_predicts_price'].abs() > fixed_results['price_predicts_flow'].abs()).sum()
    
    print(f"Coins where price predicts flow > flow predicts price: {fomo_evidence}/{len(fixed_results)}")
    print(f"Coins where flow predicts price > price predicts flow: {info_evidence}/{len(fixed_results)}")
    
    # Momentum evidence
    positive_momentum = (fixed_results['momentum_correlation'] > 0.1).sum()
    print(f"Coins with strong momentum correlation (>0.1): {positive_momentum}/{len(fixed_results)}")
    
    if positive_momentum > len(fixed_results) / 2:
        print("→ STRONG EVIDENCE FOR FOMO HYPOTHESIS")
    elif fomo_evidence > info_evidence:
        print("→ MODERATE EVIDENCE FOR FOMO HYPOTHESIS") 
    else:
        print("→ EVIDENCE FOR INFORMATION HYPOTHESIS")
    
    # Visualize
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Lead-lag scatter
    axes[0].scatter(fixed_results['price_predicts_flow'], 
                   fixed_results['flow_predicts_price'], 
                   alpha=0.7, s=100)
    axes[0].plot([-0.3, 0.3], [-0.3, 0.3], 'r--', alpha=0.5, label='Equal')
    axes[0].set_xlabel('Price Predicts Flow (FOMO)')
    axes[0].set_ylabel('Flow Predicts Price (Information)')
    axes[0].set_title('Lead-Lag Analysis (Fixed)')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # Add labels
    for _, row in fixed_results.iterrows():
        axes[0].annotate(row['coin_name'], 
                        (row['price_predicts_flow'], row['flow_predicts_price']),
                        xytext=(3, 3), textcoords='offset points', fontsize=8)
    
    # Momentum correlation
    axes[1].bar(range(len(fixed_results)), fixed_results['momentum_correlation'])
    axes[1].set_xlabel('Coin')
    axes[1].set_ylabel('Momentum Correlation')
    axes[1].set_title('Price Move → Volume Surge')
    axes[1].axhline(y=0, color='red', linestyle='--', alpha=0.7)
    axes[1].set_xticks(range(len(fixed_results)))
    axes[1].set_xticklabels(fixed_results['coin_name'], rotation=45)
    
    plt.tight_layout()
    plt.show()
    
else:
    print("No data available for analysis")

print(f"\n=== VERIFICATION ===")
print("This analysis should show DIFFERENT values for 'flow_predicts_price' and 'price_predicts_flow'")
print("If they're still identical, there's a deeper data or methodology issue.")
```

```python
# PREDICTIVE POWER ANALYSIS: Testing Signal Effectiveness for Forward Price Prediction
print("=== PREDICTIVE POWER ANALYSIS ===")
print("Testing whether discovered block and flow signals can predict forward price movements")
print()

def calculate_forward_returns(df_coin, horizons_minutes=[1, 5, 15, 30]):
    """
    Calculate forward returns at multiple time horizons
    """
    df_coin = df_coin.sort_values('block_timestamp').copy()
    
    # Calculate forward returns for each horizon
    for horizon in horizons_minutes:
        horizon_seconds = horizon * 60
        
        # Forward price at horizon
        df_coin[f'price_{horizon}min_forward'] = np.nan
        df_coin[f'return_{horizon}min_forward'] = np.nan
        
        for i in range(len(df_coin)):
            current_time = df_coin.iloc[i]['block_timestamp']
            target_time = current_time + pd.Timedelta(seconds=horizon_seconds)
            
            # Find closest price observation after target time
            future_data = df_coin[df_coin['block_timestamp'] >= target_time]
            if len(future_data) > 0:
                future_price = future_data.iloc[0]['implicit_price']
                current_price = df_coin.iloc[i]['implicit_price']
                
                if not pd.isna(future_price) and not pd.isna(current_price) and current_price > 0:
                    df_coin.iloc[i, df_coin.columns.get_loc(f'price_{horizon}min_forward')] = future_price
                    df_coin.iloc[i, df_coin.columns.get_loc(f'return_{horizon}min_forward')] = (future_price - current_price) / current_price
    
    return df_coin

def generate_block_signals(df_blocks):
    """
    Generate tradeable signals from block-level patterns
    """
    
    signals_data = []
    
    # Analyze each block for signals
    for block_id in df_blocks['block_id'].unique():
        block_data = df_blocks[df_blocks['block_id'] == block_id].sort_values('block_timestamp')
        
        if len(block_data) < 3 or block_data['sol_amount'].sum() == 0:
            continue
            
        # Get first valid price in block for future return calculation
        valid_prices = block_data[block_data['implicit_price'].notna()]
        if len(valid_prices) == 0:
            continue
            
        # Block characteristics
        block_size = len(block_data)
        total_volume = block_data['sol_amount'].sum()
        buy_count = block_data['is_buy'].sum()
        sell_count = block_data['is_sell'].sum()
        
        # Signal 1: Flow imbalance
        flow_imbalance = (buy_count - sell_count) / block_size if block_size > 0 else 0
        
        # Signal 2: Volume imbalance
        buy_volume = block_data[block_data['is_buy']]['sol_amount'].sum()
        sell_volume = block_data[block_data['is_sell']]['sol_amount'].sum()
        volume_imbalance = (buy_volume - sell_volume) / total_volume if total_volume > 0 else 0
        
        # Signal 3: Clustering patterns
        directions = ['B' if is_buy else 'S' for is_buy in block_data['is_buy']]
        direction_sequence = ''.join(directions)
        
        clustered_buys = 1 if 'BBB' in direction_sequence else 0
        clustered_sells = 1 if 'SSS' in direction_sequence else 0
        momentum_buys = 1 if direction_sequence.endswith('BB') or direction_sequence.endswith('BBB') else 0
        momentum_sells = 1 if direction_sequence.endswith('SS') or direction_sequence.endswith('SSS') else 0
        
        # Signal 4: Size concentration
        max_trade = block_data['sol_amount'].max()
        size_concentration = max_trade / total_volume if total_volume > 0 else 0
        
        # Signal 5: Trader diversity
        unique_traders = block_data['swapper'].nunique()
        trader_concentration = unique_traders / block_size if block_size > 0 else 0
        
        # Signal 6: Large vs small order imbalance
        large_threshold = block_data['sol_amount'].quantile(0.8)
        large_orders = block_data[block_data['sol_amount'] >= large_threshold]
        if len(large_orders) > 0:
            large_buy_ratio = large_orders['is_buy'].mean()
        else:
            large_buy_ratio = 0.5
            
        # Compile signals
        signal_data = {
            'coin_name': block_data['coin_name'].iloc[0],
            'block_id': block_id,
            'block_timestamp': block_data['block_timestamp'].iloc[0],
            'block_size': block_size,
            'total_volume': total_volume,
            'implicit_price': valid_prices.iloc[0]['implicit_price'],
            
            # Directional signals
            'flow_imbalance': flow_imbalance,
            'volume_imbalance': volume_imbalance,
            'buy_ratio': buy_count / block_size if block_size > 0 else 0.5,
            
            # Pattern signals
            'clustered_buys': clustered_buys,
            'clustered_sells': clustered_sells,
            'momentum_buys': momentum_buys,
            'momentum_sells': momentum_sells,
            
            # Microstructure signals
            'size_concentration': size_concentration,
            'trader_concentration': trader_concentration,
            'large_buy_ratio': large_buy_ratio,
            
            # Combined signals
            'strong_buy_signal': 1 if (flow_imbalance > 0.3 and clustered_buys == 1) else 0,
            'strong_sell_signal': 1 if (flow_imbalance < -0.3 and clustered_sells == 1) else 0,
            'momentum_signal': momentum_buys - momentum_sells,  # -1, 0, or 1
            'imbalance_strength': abs(flow_imbalance)
        }
        
        signals_data.append(signal_data)
    
    return pd.DataFrame(signals_data)

# Generate signals for all coins
print("Generating block-level signals...")
signal_df = generate_block_signals(df_blocks)
print(f"Generated {len(signal_df)} signal observations")

# Calculate forward returns for each coin
print("Calculating forward returns...")
forward_return_data = []

for coin_name in signal_df['coin_name'].unique():
    coin_signals = signal_df[signal_df['coin_name'] == coin_name].copy()
    
    if len(coin_signals) < 10:  # Need minimum observations
        continue
        
    # Calculate forward returns
    coin_with_returns = calculate_forward_returns(coin_signals, horizons_minutes=[1, 5, 15, 30])
    forward_return_data.append(coin_with_returns)

if forward_return_data:
    full_signal_df = pd.concat(forward_return_data, ignore_index=True)
    
    print(f"Forward return calculation complete: {len(full_signal_df)} observations")
    
    # Show data sample
    print("\n=== SIGNAL DATA SAMPLE ===")
    sample_cols = ['coin_name', 'block_timestamp', 'flow_imbalance', 'clustered_buys', 'strong_buy_signal', 'return_5min_forward']
    print(full_signal_df[sample_cols].head(10))
    
else:
    print("Insufficient data for forward return analysis")
    full_signal_df = pd.DataFrame()
```

```python
# CORRELATION ANALYSIS: Test Signal-Return Relationships
if not full_signal_df.empty:
    print("=== CORRELATION ANALYSIS: SIGNALS vs FORWARD RETURNS ===")
    
    # Define signal columns to test
    signal_columns = [
        'flow_imbalance', 'volume_imbalance', 'buy_ratio',
        'clustered_buys', 'clustered_sells', 'momentum_buys', 'momentum_sells',
        'size_concentration', 'trader_concentration', 'large_buy_ratio',
        'strong_buy_signal', 'strong_sell_signal', 'momentum_signal', 'imbalance_strength'
    ]
    
    # Define return horizons
    return_horizons = ['return_1min_forward', 'return_5min_forward', 'return_15min_forward', 'return_30min_forward']
    
    # Calculate correlations
    correlation_results = []
    
    for signal in signal_columns:
        for return_col in return_horizons:
            if return_col in full_signal_df.columns:
                # Remove NaN values for correlation
                valid_data = full_signal_df[[signal, return_col]].dropna()
                
                if len(valid_data) > 10:
                    corr = valid_data[signal].corr(valid_data[return_col])
                    
                    # Statistical significance test
                    from scipy.stats import pearsonr
                    stat, p_value = pearsonr(valid_data[signal], valid_data[return_col])
                    
                    correlation_results.append({
                        'signal': signal,
                        'return_horizon': return_col,
                        'correlation': corr,
                        'p_value': p_value,
                        'n_observations': len(valid_data),
                        'significant': p_value < 0.05
                    })
    
    corr_df = pd.DataFrame(correlation_results)
    
    # Show best correlations
    print("\\n=== TOP SIGNAL-RETURN CORRELATIONS ===")
    top_correlations = corr_df.nlargest(15, 'correlation')
    for _, row in top_correlations.iterrows():
        significance = "***" if row['p_value'] < 0.001 else "**" if row['p_value'] < 0.01 else "*" if row['p_value'] < 0.05 else ""
        print(f"{row['signal']:20s} → {row['return_horizon']:18s}: {row['correlation']:7.4f} (p={row['p_value']:.3f}) {significance} (n={row['n_observations']})")
    
    print("\\n=== STRONGEST NEGATIVE CORRELATIONS ===")
    bottom_correlations = corr_df.nsmallest(10, 'correlation')
    for _, row in bottom_correlations.iterrows():
        significance = "***" if row['p_value'] < 0.001 else "**" if row['p_value'] < 0.01 else "*" if row['p_value'] < 0.05 else ""
        print(f"{row['signal']:20s} → {row['return_horizon']:18s}: {row['correlation']:7.4f} (p={row['p_value']:.3f}) {significance} (n={row['n_observations']})")
    
    # Count significant relationships
    significant_correlations = corr_df[corr_df['significant'] == True]
    print(f"\\n=== STATISTICAL SIGNIFICANCE SUMMARY ===")
    print(f"Total signal-return combinations tested: {len(corr_df)}")
    print(f"Statistically significant relationships (p<0.05): {len(significant_correlations)} ({len(significant_correlations)/len(corr_df):.1%})")
    print(f"Strong correlations (|r| > 0.1): {(abs(corr_df['correlation']) > 0.1).sum()}")
    print(f"Very strong correlations (|r| > 0.2): {(abs(corr_df['correlation']) > 0.2).sum()}")
    
    # Breakdown by return horizon
    print(f"\\n=== SIGNIFICANCE BY TIME HORIZON ===")
    for horizon in return_horizons:
        horizon_data = corr_df[corr_df['return_horizon'] == horizon]
        significant_count = (horizon_data['significant'] == True).sum()
        strong_count = (abs(horizon_data['correlation']) > 0.1).sum()
        print(f"{horizon:20s}: {significant_count:2d}/{len(horizon_data):2d} significant ({significant_count/len(horizon_data):.1%}), {strong_count} strong correlations")
    
    # Best signals by category
    print(f"\\n=== BEST SIGNALS BY CATEGORY ===")
    
    # Directional signals
    directional_signals = ['flow_imbalance', 'volume_imbalance', 'buy_ratio']
    directional_corrs = corr_df[corr_df['signal'].isin(directional_signals)]
    if not directional_corrs.empty:
        best_directional = directional_corrs.loc[directional_corrs['correlation'].abs().idxmax()]
        print(f"Best Directional Signal: {best_directional['signal']} → {best_directional['return_horizon']} (r={best_directional['correlation']:.4f})")
    
    # Pattern signals
    pattern_signals = ['clustered_buys', 'clustered_sells', 'momentum_buys', 'momentum_sells']
    pattern_corrs = corr_df[corr_df['signal'].isin(pattern_signals)]
    if not pattern_corrs.empty:
        best_pattern = pattern_corrs.loc[pattern_corrs['correlation'].abs().idxmax()]
        print(f"Best Pattern Signal: {best_pattern['signal']} → {best_pattern['return_horizon']} (r={best_pattern['correlation']:.4f})")
    
    # Combined signals
    combined_signals = ['strong_buy_signal', 'strong_sell_signal', 'momentum_signal']
    combined_corrs = corr_df[corr_df['signal'].isin(combined_signals)]
    if not combined_corrs.empty:
        best_combined = combined_corrs.loc[combined_corrs['correlation'].abs().idxmax()]
        print(f"Best Combined Signal: {best_combined['signal']} → {best_combined['return_horizon']} (r={best_combined['correlation']:.4f})")
    
else:
    print("No data available for correlation analysis")
```

```python
# PRACTICAL TRADING ANALYSIS: Signal Performance & Strategy Testing
if not full_signal_df.empty and 'corr_df' in locals():
    print("=== PRACTICAL TRADING ANALYSIS ===")
    
    def analyze_signal_performance(df, signal_col, return_col, signal_threshold=None):
        """
        Analyze trading performance for a specific signal
        """
        df_clean = df[[signal_col, return_col]].dropna()
        
        if len(df_clean) == 0:
            return None
            
        # Define signal conditions
        if signal_threshold is None:
            if signal_col in ['strong_buy_signal', 'strong_sell_signal', 'clustered_buys', 'clustered_sells', 'momentum_buys', 'momentum_sells']:
                # Binary signals
                buy_condition = df_clean[signal_col] == 1
                sell_condition = df_clean[signal_col] == 0  # For binary, opposite is neutral
            else:
                # Continuous signals - use quantiles
                q75 = df_clean[signal_col].quantile(0.75)
                q25 = df_clean[signal_col].quantile(0.25)
                buy_condition = df_clean[signal_col] >= q75
                sell_condition = df_clean[signal_col] <= q25
        else:
            buy_condition = df_clean[signal_col] >= signal_threshold
            sell_condition = df_clean[signal_col] <= -signal_threshold
        
        # Calculate returns for each condition
        buy_returns = df_clean[buy_condition][return_col]
        sell_returns = df_clean[sell_condition][return_col]
        neutral_returns = df_clean[~(buy_condition | sell_condition)][return_col]
        
        results = {
            'signal': signal_col,
            'return_horizon': return_col,
            'total_observations': len(df_clean),
            
            # Buy signals
            'buy_signals': len(buy_returns),
            'buy_hit_rate': (buy_returns > 0).mean() if len(buy_returns) > 0 else 0,
            'buy_avg_return': buy_returns.mean() if len(buy_returns) > 0 else 0,
            'buy_return_std': buy_returns.std() if len(buy_returns) > 0 else 0,
            
            # Sell signals  
            'sell_signals': len(sell_returns),
            'sell_hit_rate': (sell_returns < 0).mean() if len(sell_returns) > 0 else 0,  # Correct direction for sells
            'sell_avg_return': sell_returns.mean() if len(sell_returns) > 0 else 0,
            'sell_return_std': sell_returns.std() if len(sell_returns) > 0 else 0,
            
            # Neutral/baseline
            'neutral_signals': len(neutral_returns),
            'neutral_avg_return': neutral_returns.mean() if len(neutral_returns) > 0 else 0,
            
            # Performance metrics
            'buy_sharpe': (buy_returns.mean() / buy_returns.std()) if len(buy_returns) > 0 and buy_returns.std() > 0 else 0,
            'sell_sharpe': (abs(sell_returns.mean()) / sell_returns.std()) if len(sell_returns) > 0 and sell_returns.std() > 0 else 0,
        }
        
        return results
    
    # Test best performing signals
    print("\\n=== SIGNAL PERFORMANCE ANALYSIS ===")
    
    # Get top 10 absolute correlations for detailed analysis
    top_signals = corr_df.nlargest(10, 'correlation')
    
    performance_results = []
    
    for _, row in top_signals.iterrows():
        performance = analyze_signal_performance(
            full_signal_df, 
            row['signal'], 
            row['return_horizon']
        )
        if performance:
            performance_results.append(performance)
    
    # Display results
    print(f"\\nTop Signal Performance (Buy Signals):")
    print(f"{'Signal':<20} {'Horizon':<18} {'Count':<6} {'HitRate':<8} {'AvgRet':<8} {'Sharpe':<7}")
    print("-" * 75)
    
    for result in performance_results:
        if result['buy_signals'] >= 5:  # Only show signals with sufficient data
            print(f"{result['signal']:<20} {result['return_horizon']:<18} {result['buy_signals']:<6} "
                  f"{result['buy_hit_rate']:<8.1%} {result['buy_avg_return']:<8.4f} {result['buy_sharpe']:<7.2f}")
    
    # Calculate strategy-level performance
    print(f"\\n=== STRATEGY-LEVEL PERFORMANCE ===")
    
    # Simple strategy: Use best signal for each time horizon
    strategy_results = {}
    
    for horizon in return_horizons:
        if horizon in full_signal_df.columns:
            horizon_signals = corr_df[corr_df['return_horizon'] == horizon].copy()
            
            if len(horizon_signals) > 0:
                # Find best absolute correlation
                best_signal_row = horizon_signals.loc[horizon_signals['correlation'].abs().idxmax()]
                
                # Test this signal's performance
                strategy_performance = analyze_signal_performance(
                    full_signal_df,
                    best_signal_row['signal'],
                    horizon
                )
                
                if strategy_performance:
                    strategy_results[horizon] = {
                        'best_signal': best_signal_row['signal'],
                        'correlation': best_signal_row['correlation'],
                        'performance': strategy_performance
                    }
    
    print("Best Signal Strategy by Horizon:")
    print(f"{'Horizon':<18} {'Signal':<20} {'Corr':<7} {'Buy Signals':<11} {'Hit Rate':<9} {'Avg Return':<11}")
    print("-" * 85)
    
    for horizon, data in strategy_results.items():
        perf = data['performance']
        if perf['buy_signals'] >= 3:
            print(f"{horizon:<18} {data['best_signal']:<20} {data['correlation']:<7.3f} "
                  f"{perf['buy_signals']:<11} {perf['buy_hit_rate']:<9.1%} {perf['buy_avg_return']:<11.4f}")
    
    # Visualization
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # 1. Correlation heatmap
    pivot_corr = corr_df.pivot(index='signal', columns='return_horizon', values='correlation')
    sns.heatmap(pivot_corr, annot=True, cmap='RdBu_r', center=0, ax=axes[0,0])
    axes[0,0].set_title('Signal-Return Correlations Heatmap')
    axes[0,0].set_xlabel('Return Horizon')
    axes[0,0].set_ylabel('Signal Type')
    
    # 2. Significance scatter plot
    axes[0,1].scatter(corr_df['correlation'], -np.log10(corr_df['p_value']), 
                     alpha=0.6, c=['red' if x else 'blue' for x in corr_df['significant']])
    axes[0,1].axhline(y=-np.log10(0.05), color='red', linestyle='--', alpha=0.5, label='p = 0.05')
    axes[0,1].set_xlabel('Correlation')
    axes[0,1].set_ylabel('-log10(p-value)')
    axes[0,1].set_title('Statistical Significance vs Correlation')
    axes[0,1].legend()
    
    # 3. Signal strength distribution
    if 'imbalance_strength' in full_signal_df.columns:
        axes[1,0].hist(full_signal_df['imbalance_strength'], bins=30, alpha=0.7, edgecolor='black')
        axes[1,0].set_xlabel('Flow Imbalance Strength')
        axes[1,0].set_ylabel('Frequency')
        axes[1,0].set_title('Distribution of Signal Strength')
    
    # 4. Best signal performance by horizon
    if strategy_results:
        horizons = list(strategy_results.keys())
        hit_rates = [strategy_results[h]['performance']['buy_hit_rate'] for h in horizons]
        avg_returns = [strategy_results[h]['performance']['buy_avg_return'] for h in horizons]
        
        ax2 = axes[1,1]
        ax2.bar(range(len(horizons)), hit_rates, alpha=0.7, label='Hit Rate')
        ax2.set_ylabel('Hit Rate', color='blue')
        ax2.set_ylim(0, 1)
        
        ax3 = ax2.twinx()
        ax3.plot(range(len(horizons)), avg_returns, 'ro-', label='Avg Return')
        ax3.set_ylabel('Average Return', color='red')
        
        ax2.set_xticks(range(len(horizons)))
        ax2.set_xticklabels([h.replace('return_', '').replace('_forward', '') for h in horizons])
        ax2.set_xlabel('Time Horizon')
        ax2.set_title('Strategy Performance by Horizon')
        
        # Add legends
        lines1, labels1 = ax2.get_legend_handles_labels()
        lines2, labels2 = ax3.get_legend_handles_labels()
        ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("No data available for trading analysis")
```
